```python
import decimal
ctx = decimal.getcontext()
ctx
```

```python
ctx.prec,ctx.rounding
```

```python
# change the context globall
ctx.prec = 10
```

```python
with decimal.localcontext() as ctx:
    # only change local context
    ctx.prec = 2
    ctx.rounding = decimal.ROUND_HALF_UP
```

```python
x = decimal.Decimal('1.25')
y = decimal.Decimal('1.35')

print(round(x,1))
print(round(y,1))
```
