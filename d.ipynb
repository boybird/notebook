```python
import numpy as np
import pandas as pd
import os
os.getcwd()
```

```python
def get_data():
    df = pd.read_csv("./data/ecommerce_data.csv")
    data = df.values
    
    X = data[:, : -1]
    Y = data[:, -1]
    
    X[:,1] = (X[:,1] - X[:,1].mean() ) / X[:,1].std()
    X[:,2] = (X[:,2] - X[:,2].mean() ) / X[:,2].std()
    
    N, D = X.shape
    X2 = np.zeros((N,D+3))
    X2[:, 0:(D-1)] = X[:, 0:(D-1)]
    
    for n in range(N ):
        t = int(X[n, D-1])
        X2[n,t+D-1] = 1
    
    Z = np.zeros((N,4))
    Z[np.arange(N),X[:,D-1].astype(np.int32)] = 1
    assert(np.abs(X2[:,-4:] -Z).sum()< 10e-10)
    return X2,Y

def get_binaray_data():
    X,Y = get_data()
    X2 = X[Y<=1]
    Y2 = Y[Y<=1]
    return X2,Y2
```

```python
X,Y = get_binaray_data()
```

```python
D = X.shape[1]
W = np.random.randn(D)
b = 0
```

```python
def sigmod(a):
    return 1/(a+np.exp(-1))

def forward(X,W,b):
    return sigmod(X.dot(W)+b)
```

```python
P_Y_given_X = forward(X,W,b)
P_Y_given_X
```

```python
predictions = np.round(P_Y_given_X)
predictions
```

```python
def classifyication_rates(Y,P):
    return np.mean(Y==P)
```

```python
print( "Scores", classifyication_rates(Y,predictions))

```
