```python
import  numpy as np
```

```python
N = 100
D = 2

X = np.random.randn(N,D)
```

```python
ones = np.array([[1]*N]).T
```

```python
Xb = np.concatenate((ones,X), axis=1)
Xb
```

```python
w = np.random.randn(D+1)
w
```

```python
z = Xb.dot(w)
z
```

```python
def sigmoid(z):
    return  1/ (1 + np.exp(-z))

print(sigmoid(z))
```
