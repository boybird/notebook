```python
def common_keys(d1,d2):
    keys = d1.keys() & d2.keys()
    d = { k: (d1[k],d2[k])
        for k in keys
    }
    return d
```

```python
d1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
d2 = {'b': 20, 'c': 30, 'y': 40, 'z': 50}
common_keys(d1,d2)

# common_keys(d1,d2)
```

```python
def combined_dicts(*dicts):
    unsorted = {}
    for d in dicts:
        for k ,v in d.items():
            # unsorted[k] = unsorted[k]+v if k in unsorted else v
            unsorted[k] = unsorted.get(k,0) + v
    
    return dict(sorted(unsorted.items(), key=lambda a:a[1],reverse=True))

combined_dicts(d1,d2,d3)
```
